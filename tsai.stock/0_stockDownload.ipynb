```python
#download stock data using fdr library
KospiCodeList = fdr.StockListing('KOSPI')["Symbol"].tolist() 
KosdaqCodeList = fdr.StockListing('KOSDAQ')["Symbol"].tolist() 
fdr_tickers=KospiCodeList+KosdaqCodeList
print(len(fdr_tickers))
print(fdr_tickers[:10])


for code in tqdm(fdr_tickers):
    PATH=Path('./data/fdr/')/(code+'.csv')
    PATH.parent.mkdir(parents=True, exist_ok=True)

    try:        
        df = fdr.DataReader(code, '2000-01-01', '2021-07-01')
        if(len(df)==0): continue
        tempDF=pd.DataFrame()
        tempDF[['Close','Open','High','Low','Volume']]=df[['Close','Open','High','Low','Volume']]
        tempDF.to_csv(PATH) 
        time.sleep(1)
    
    except Exception as e:   
        print( e)
        print(code)
```

```python
#download stock data using pykrx library
dateEnd="20220701"
KosdaqTickers = stock.get_market_ticker_list(dateEnd, market="KOSDAQ")
KospiTickers = stock.get_market_ticker_list(dateEnd, market="KOSPI")
pykrx_tickers=KospiTickers
print(len(pykrx_tickers))
print(pykrx_tickers[:10])


for code in tqdm(pykrx_tickers):
    PATH=Path('./data/pykrx/')/(code+'.csv')
    PATH.parent.mkdir(parents=True, exist_ok=True)
    
    try:
        df=stock.get_market_ohlcv_by_date("20000101", dateEnd, code)
        if(len(df)==0): continue
        tempDF=pd.DataFrame()
        tempDF[['Close','Open','High','Low','Volume']]=df[['종가','시가','고가','저가','거래량']]
        tempDF.to_csv(PATH) 
        time.sleep(1)
        
    except Exception as e:   
        print( e)
        print(code)
```
